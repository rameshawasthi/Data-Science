##Adding Annotations

```python
x = np.linspace(0,10)

y1 = x 
y2 = 8-x

fig, ax = plt.subplots()
plt.plot(x,y1,label='supply')
plt.plot(x,y2,label="demand")

ax.annotate("Equilibrium", xy=(4,4), xytext=(3,2),\
            fontsize=12, fontweight="semibold", \
             arrowprops = dict(linewidth=2, arrowstyle="->"))

plt.xlabel('quantity', fontsize=12)
plt.ylabel('price',fontsize=12)

plt.legend()
plt.show()
```

```python
x = np.linspace(0,10)

y1 = x 
y2 = 8-x
#plot the data
fig, ax = plt.subplots()
plt.plot(x,y1,label='supply')
plt.plot(x,y2,label="demand")
#Annotate the equilibrium point with arrow and text
bbox_props = dict(boxstyle="rarrow", fc=(0.8,0.9,0.9), lw=2)
t  = ax.text(2,4, "Equilibrium", ha="center", va="center", rotation = 0,
             size=10, bbox=bbox_props)
#label the axes
plt.xlabel('quantity', fontsize=12)
plt.ylabel('price',fontsize=12)

plt.legend()
plt.show()
```

```python
#creating numpy array
intergers = np.arange(1, 101)
print('intergers   :', *intergers)

#creating own function
def modulo(val):
  return(val % 10)
#adding into numpy
mod_10 = np.frompyfunc(modulo, 1, 1)

#using function over numpy array
mod_intergers = mod_10(intergers)
print("mod_intergers :", *mod_intergers)
```

```python
matrix1 = np.matrix([[1,2],[3,4]])
matrix1
```

```python
eigenvalues, eigenvectors = np.linalg.eig(matrix1)
print("Eigenvalues: ", eigenvalues)
print("Eigenvectors: ", eigenvectors)
```

** Application : Linear Regression

* One of the most important and widely used models for machine learning algorithms.

House Market

* input: price data from 2012-2022
* output : Average house market 2023
* assume Relationship - squared
* y - ax**2 + bx + c

```python
year = np.arange(1,12)
price=  np.array([129000, 133000, 13800, 1443000,1461000,1489000,1490000,1512000,1519000,1540000,1590000])
year
```

```python
x, y, w = np.polyfit(year, price, 2)

print("x:",x)
print("y:",y)
print("w:",w)
```

```python
print("Estimated price fo r2022:", x*11**2 + y*11+w)
```

```python
plt.plot(year, price, color = "blue")
plt.scatter(year, price, color = "blue")
plt.scatter(11, x*11**2 + y*11 + w, color = 'red')
plt.title("Estimated House Price")
plt.xlabel('year')
plt.ylabel('price')
```
